```python
import pandas as pd
import requests
import json
```

```python
key = open("eia_api_key.txt", "r").readlines()[0]
```

```python
branch = "electricity"
url = f"https://api.eia.gov/v2/{branch}/?api_key="
r = requests.get(url + key)
electricity = json.loads(r.text)
```

```python
branch = "petroleum"
url = f"https://api.eia.gov/v2/{branch}/?api_key="
r = requests.get(url + key)
petroleum = json.loads(r.text)
```
